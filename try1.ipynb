```python
x = 0
def data_cleaner(dirty_data):
    latitudes = []
    new_df = df[['business_address', 'category_name','business_category_name','followers','is_verified','posts_count','posts','avg_engagement']]
    new_df = new_df.dropna(subset="business_address")
    latitudes = []
    longitudes = []
    # Iterate through each value in the 'business_address' column after dropping NaN values
    for value in df['business_address'].dropna():
        # Check if the value is a dictionary
        if isinstance(value, dict):
            # Check if 'lat' key exists and is not None
            if 'lat' in value and value['lat'] is not None:
                latitudes.append(value['lat'])
            else:
                latitudes.append(None)
        else:
            latitudes.append(None)
    # Same for longitudes
    for value in df['business_address'].dropna():
        # Check if the value is a dictionary
        if isinstance(value, dict):
            # Check if 'long' key exists and is not None
            if 'long' in value and value['long'] is not None:
                longitudes.append(value['long'])
            else:
                longitudes.append(None)
        else:
            longitudes.append(None)
    new_df["latitudes"] = latitudes
    new_df["longitudes"] = longitudes
    new_df = new_df.drop(columns = "business_address")
    new_df = new_df.dropna(subset = ['latitudes', 'longitudes'])
    x = len(latitudes)
    return new_df
```

```python
final_df = data_cleaner(df)
```
